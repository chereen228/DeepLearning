```python
data= pd.read_excel("crawl/naver_cafe_ssal_contents_2019_전처리완.xlsx")
data=data.dropna(axis=0, how='any', subset = ['content']) 
data.info()
data.reset_index(drop=True, inplace=True)
data
```

```python
conlist=[]
for i in range(len(data2)):

    a=[]
    a.append(data2["content"][i].split("||-||"))
    
    d=pd.DataFrame(a).T
    d=d[d[0]!=""].reset_index()
    d=d[d[0]!="구요"].reset_index()
    conca1=d[d[0].str.contains("이유식|아이|아기|애기|아가|미음|분유",na=False)]
    conidex=conca1.index
    list(conidex)


    for i in list(conidex):
        try:

             conlist.append(str(d[0][i-1])+" "+str(d[0][i])+" "+str(d[0][i+1]))

        except:
            try:
                conlist.append(str(d[0][i])+" "+str(d[0][i+1]))
            except:
                try:
                    conlist.append(str(d[0][i-1])+" "+str(d[0][i]))
                except:
                    conlist.append(str(d[0][i]))
```

```python
aaaa=pd.DataFrame(conlist)
aaaa
```
