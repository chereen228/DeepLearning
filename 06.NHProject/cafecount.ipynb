```python
total2 = pd.read_excel('crawl/2019_cafe_content_spelled_hanspell.xlsx')
print(total2.shape)
total2.head()
```

```python
del total2['Unnamed: 2']
```

```python
# 형태소 분석 (Hannanum)
q = " "
nouns2 = []

for i in tqdm(range(9828)):
    q += str(total2['spell_checked'][i])
    if i % 50 == 0:
        nouns = hannanum.nouns(q)
        # print(nouns)
        nouns2.append(nouns)
        q = " "
```

```python
answer = sum(nouns2, [])
answer
```

```python
count = Counter(answer)
remove_char_counter = Counter({x: count[x] for x in count if len(x) > 1})

print(count)
```

```python
words = dict(count.most_common())
words
```

```python
df = pd.DataFrame(list(words.items()), columns=['word', 'cnt'])
df
```

```python
df.to_csv('2019cafe_count_hannanum.csv', encoding='utf-8-sig', index=False)
```
