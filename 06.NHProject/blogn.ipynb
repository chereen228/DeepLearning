```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
import numpy as np 
import pandas as pd 


browser = webdriver.Chrome('./chromedriver') 


#########아래 주소만 수정하면 됩니다
#cnt는 게시물 순서를 나타냅니다 
#혹시 중간에 멈추게 된다면 처음부터 다시 돌리지 마시고 해당 데이터를 파일로 저장하시고 그 다음 순서 숫자를 cnt에 입력해주세요 
#ex) 만약 데이터가 336개 까지 되었다면 해당 데이터는 저장하고 cnt=0 -> cnt=336 으로 해서 336 이후 데이터를 받아오시고 해당 데이터를 파일로 저장후 합치면 됩니다(엑셀에서 수작업 가능(쉬움))
#(데이터 확인을 위해 앞 몇개 순서로 하면 좋습니다 위 사례로 예를 들면 cnt=330 을 하시면 데이터 앞 확인 후 데이터 합치는게 가능합니다(선택사항))
browser.get("https://search.naver.com/search.naver?where=blog&query=%EC%8C%80%20-%EC%97%85%EC%B2%B4%2C%20-%EC%A0%9C%EA%B3%B5%EB%B0%9B%2C%20-%EC%A7%80%EC%9B%90%EB%B0%9B%2C%20-%EC%9B%90%EA%B3%A0%EB%A3%8C%2C%20-%EB%AC%B4%EC%83%81%2C%20-%ED%98%91%EC%B0%AC%2C%20-%EC%86%8C%EC%A0%95%EC%9D%98%2C%20-%EC%82%AC%EB%A3%8C%2C%20-%EA%B3%A0%EB%9E%98%EB%B0%A5%2C%20-%EA%B8%B8%EB%83%A5%EC%9D%B4%2C%20-%EA%B3%A0%EC%96%91%EC%9D%B4%2C%20-%EA%B0%95%EC%95%84%EC%A7%80&sm=tab_opt&nso=so%3Ar%2Cp%3Afrom20190601to20190630")
time.sleep(1)


#해당 코드는 스크롤을 끝까지 내려주는 코드입니다 혹시 끝까지 안내려가면 숫자를 늘려주시면 됩니다(range(1030)에서 1030숫자 건들면 됩니다)
while True:
    for w in range(1030):
        browser.find_element_by_css_selector("html").send_keys(Keys.END)
    break


##최종블로그
titles = []
dates = []
#내용
contents = []
#id
ids = []
#댓글
chats = []
chats2 = []


title = browser.find_elements_by_css_selector("a.api_txt_lines.total_tit")
date = browser.find_elements_by_css_selector("span.sub_time.sub_txt")

cnt = 0
while True:
    try:

        titles.append(title[cnt].text)
        dates.append(date[cnt].text)

        title[cnt].click()


        browser.switch_to.window(browser.window_handles[-1])

        try:

            browser.switch_to.frame('mainFrame')


            #content


            try:

                blogcontent = browser.find_elements_by_css_selector(
                    "div.se-main-container")
                for i in blogcontent:
                    contents.append(i.text)

                if blogcontent == []:
                    blogcontent = browser.find_elements_by_css_selector(
                        "div#postViewArea")
                    for i in blogcontent:
                        contents.append(i.text)


                if blogcontent == []:
                    contents.append(np.nan)

                
            except:
                contents.append(np.nan)


            #id


            try:        
                ids.append(browser.find_element_by_css_selector(
                    "span.itemfont.col").text)
            except:
                ids.append(np.nan)


            #chat

            try:
                browser.find_element_by_css_selector(
                    "div.area_comment.pcol2").click()
                time.sleep(1)
                blogchat = browser.find_elements_by_css_selector(
                    "span.u_cbox_contents")
                midchat=[]
                q = ""

                for j in blogchat:
                    q += j.text+" "
                    midchat.append(j.text)
                chats.append(q)
                chats2.append(midchat)
            except:
                chats.append(np.nan)
                chats2.append(np.nan)
            
            browser.close()
            browser.switch_to.window(browser.window_handles[0])
        
        except:
            
            browser.close()
            browser.switch_to.window(browser.window_handles[0])

            chats.append(np.nan)
            contents.append(np.nan)
            ids.append(np.nan)


    except:
        print("END")
        break



    cnt += 1


    # if cnt % 15 == 0:
    #     browser.find_element_by_css_selector("html").send_keys(Keys.END)
```

```python
a=pd.DataFrame(titles,columns=["title"])
b = pd.DataFrame(dates, columns=["date"])

c = pd.DataFrame(chats, columns=["chat"])
d = pd.DataFrame(contents, columns=["content"])
e = pd.DataFrame(ids,columns=["id"])
```
