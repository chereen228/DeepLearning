```python
import pandas as pd
import plotly
```

```python
df1 = pd.read_csv('data/경남_노인의료복지시설_전처리.csv', encoding='utf-8')
df2 = pd.read_csv('data/경남_노인주거복지_전처리.csv', encoding='utf-8')
df3 = pd.read_csv('data/경남_복지관_전처리.csv', encoding='utf-8')
```

```python
df1.info()
df2.info()
```

```python
df3.info()
```

```python
wf = pd.concat([df1,df2])
wf
```

```python
wfare = pd.concat([wf,df3])
wfare
```

```python
wf2 = wfare.dropna(axis=0, how='any', subset=['facil'])
wf2 = wf2.reset_index()
wf2.info()
```

```python
wf2 = wfare.drop_duplicates(['facil'])
wf2 = wf2.reset_index()
wf2.info()
```

```python
wf2.to_csv('data/전체복지시설.csv', encoding='utf-8-sig')
```

```python
wfr = pd.read_csv('data/전체복지시설.csv')
wfr
```

```python
wf_top = wfr['index'].groupby(wfr['city']).count().sort_values(ascending=False)
wf_top.head()
```

```python
wf_top.to_csv('data/복지시설_시군별.csv', encoding='utf-8-sig')
```

```python
inf1 = pd.read_csv('data/경남_병원_전처리.csv', encoding='utf-8')
inf2 = pd.read_csv('data/경남_소방서_전처리.csv', encoding='utf-8')
inf3 = pd.read_csv('data/경남_응급환자이송업_전처리.csv', encoding='utf-8')
inf4 = pd.read_csv('data/경남_의원_전처리.csv', encoding='utf-8')
```
