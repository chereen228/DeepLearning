```python
import numpy as np
import pandas as pd
```

```python
hc = pd.read_csv('select/합천버스.csv')
hc.info()
hc
```

```python
hc = hc.dropna(axis=0, how='any')
hc = hc.reset_index()
hc.info()
```

```python
hc_bus = hc['index'].groupby(hc['eup']).count().sort_values(ascending=False)
hc_bus
```

```python
hc_bus.to_csv('합천_교통접근성.csv', encoding='utf-8-sig')
```
